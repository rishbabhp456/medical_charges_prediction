```python
# we can perform the above functionality in function
```

```python
def test_med_ins_pred(age,gender,bmi,children,smoker,region):
    # values conversion and handling one_encoding
    gender =  col_data['gender'][gender]
    smoker = col_data['smoker'][smoker]
    region = f"region_{region}"
    region_index = np.where(col_names == region)[0][0]
    test_array= np.zeros((1,model.n_features_in_))

    #user input values insertion
    test_array[0,0] = age
    test_array[0,1] = gender
    test_array[0,2] = bmi
    test_array[0,3] = children
    test_array[0,4] = smoker
    test_array[0,region_index]

    #for getting result
    test_df = pd.DataFrame(test_array, columns= col_names)
    pred_charges = np.around(model.predict(test_df)[0],2)

    return pred_charges
```

```python
age = 25
gender ='female'
bmi = 28.455
children = 2
smoker='no'
region = "northeast"

predict_charges = test_med_ins_pred(age,gender,bmi,children,smoker,region)
print("Predicted Medical Charges are:",predict_charges)
```

```python
age = 32
gender = "male"
bmi = 25.435
children = 2
smoker = "yes"
region = "southwest"
pred_charges = test_med_ins_pred(age, gender, bmi, children, smoker, region)
print("Predicted Medical Insurance Charges :",pred_charges)
```
